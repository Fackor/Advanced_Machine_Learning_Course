```
# Directly run luke to fine-tune.
%cd luke
!python -m examples.cli                 \
    --model-file=luke_large_500k.tar.gz \
    --output-dir=/content/train_output  \
    ner run                             \
    --data-dir=/content/upload2         \
    --train-batch-size=1                \
    --gradient-accumulation-steps=2     \
    --learning-rate=1e-5                \
    --num-train-epochs=3                \
    --max-seq-length=64                \
    --fp16
%cd ..
```
